# Lab 9: Managing a backported change

1\. First we need to get a copy of the Diffutils repository, in Git format, from the file ~eggert/src/gnu/diffutils on the SEASnet GNU/Linux servers.

```bash
git clone classtzh@lnxsrv.seas.ucla.edu:~eggert/src/gnu/diffutils
```

2\. Then we willget a log of changes to Diffutils’ master branch using the `git log` command, and put it into the file git-log.txt.

```bash
cd diffutils
```

```bash
git log | head -100
```

```bash
git log > ../git-log.txt
```

3\. Generate a list of tags used for Diffutils using the `git tag` command, and put it into the file `git-tags.txt`.

```bash
git tag
```

```bash
git tag > ../git-tags.txt
```

4\. Find the commit entitled “maint: quote 'like this' or "like this", not `like this'”, and generate a patch for that commit, putting it into the file quote-patch.txt.

```bash
git log | grep "maint: quote" -B 4
```

So we know that the commit ID of the desired commit is "62ca21c8c1a5aa3488589dcb191a4ef04ae9ed4f".

With the commit ID, we can now create the patch.

```bash
git format-patch -1 62ca21c8c1a5aa3488589dcb191a4ef04ae9ed4f
```

```bash
cp 0001-maint-quote-like-this-or-like-this-not-like-this.patch ..
```

5\. Now we are going to check out version 3.0 of the repository.

```bash
git checkout v3.0
```

6\. We will now use the `patch` command to apply quote-patch.txt to version 3.0. Some patches cannot be applied successfully and we will skip all of them.

```bash
patch -p0 < 0001-maint-quote-like-this-or-like-this-not-like-this.patch
```

Below are the files that either cannot be patched at all or have some part of the patch failed:

* HACKING
* README
* cfg.mk
* doc/diffutils.texi
* man/help2man
* diff.c
* diff3.c

Because all the failed patches produce a report with names `*.rej`, for reference we will also print out these reports.

```bash
find . -regex ".*.rej" -exec echo {} \; -exec cat {} \; -exec echo \;
```

After some examinations, we found that the majority of the patches failed because the lines they intend to change are not in the specified line number any more.

7\. We will now use the git status command to get an overview of the changes in the repository.

```bash
cd diffutils
```

```bash
git status
```

Quite a few files are modified and some more are created.

8\. 9\. Because here we only need to change the left quote symbol in the C source code, we are going to revert all the other changes with the `vc-diff` mode of `Emacs`. 

First we will revert all the changes made to files other than C source code. After completion, we can see that now only C source codes are modified.

```bash
git status
```

Then we are now going to go through the changes in the C files and revert all the changes made other than the left quote symbol. We will do this with the help of Emacs' vc-diff mode.

We walked through all the changed made to the tracked C files very carefully and did not find any undesired change.

10\. Now we are going to examine the files `src/*.c.rej` and see if there is any desired change not correctly patched.

```bash
git status | grep "src/.*.c.rej"
```

There are only two such files, and upon examinations, the patches are not successfully applied because the lines they intend to change moved in locations. We thus manually changed all the \` into ".

11\. Since we don’t plan on adding any files in your patch, we will proceed to remove all untracked files that git status warns  us about.

```bash
git clean -f
```

12\. Now there should be no untracked files remaining and the only modified files should be C source codes. We are going to verify it with `git status` and `git diff`.

```bash
git status
```

```bash
git diff
```

We are also going to use `git diff` to generate a patch file.

```bash
git diff > ../quote-3.0-patch.txt
```
